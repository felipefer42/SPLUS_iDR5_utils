```python
# Current (iDR5)
#http://svo2.cab.inta-csic.es/svo/theory/fps3/index.php?mode=browse
_lambda_ref = {  'SPLUS_U': 3533.29, 'SPLUS_F378': 3773.13,
              'SPLUS_F395': 3940.70, 'SPLUS_F410': 4095.27,
              'SPLUS_F430': 4292.39,    'SPLUS_G': 4758.49,
              'SPLUS_F515': 5133.15,    'SPLUS_R': 6251.83,
              'SPLUS_F660': 6613.88,    'SPLUS_I': 7670.59,
              'SPLUS_F861': 8607.59,    'SPLUS_Z': 8936.64,
               'GALEX_FUV': 1535.08,  'GALEX_NUV': 2300.78,
                  'SDSS_U': 3556.52,     'SDSS_G': 4702.50,
                  'SDSS_R': 6175.58,     'SDSS_I': 7489.98,
                  'SDSS_Z': 8946.71,       'PS_G': 4849.11,
                    'PS_R': 6201.20,       'PS_I': 7534.96,
                    'PS_Z': 8674.20,    'GAIA_BP': 5048.62,
                  'GAIA_G': 6246.76,    'GAIA_RP': 7740.87,
                    'SM_U': 3493.36,       'SM_V': 3835.93,
                    'SM_G': 5075.19,       'SM_R': 6138.44,
                    'SM_I': 7767.98,       'SM_Z': 9145.99,
                 'GAIA3_G': 6217.59,   'GAIA3_BP': 5109.71,
                'GAIA3_RP': 7769.02}

# Plot synthetic magnitudes
def wavelength_to_hex(wavelength, gamma=0.8):
    '''This converts a given wavelength of light to an
    approximate hex color value. The wavelength must be given
    in angstrons in the range from 3000 A through 10000 A

    Based on code by Dan Bruton
    http://www.physics.sfasu.edu/astro/color/spectra.html
    '''

    wavelength = float(wavelength) / 10.

    if 200 <= wavelength < 350:
        attenuation = 0.5 + 0.5 * (wavelength - 200) / (440 - 200)
        R = ((-(wavelength - 440) / (440 - 200)) * attenuation) ** gamma
        G = 0.0
        B = (1.0 * attenuation) ** gamma
    elif 350 <= wavelength <= 380:
        attenuation = 0.8 + 0.2 * (wavelength - 350) / (430 - 350)
        R = ((-(wavelength - 430) / (430 - 350)) * attenuation) ** gamma
        G = 0.2*(wavelength - 350) / (380 - 350) 
        B = (1.0 * attenuation) ** gamma
    elif 380 <= wavelength <= 430:
        attenuation = 0.3 + 0.7 * (wavelength - 350) / (430 - 350)
        R = ((-(wavelength - 430) / (430 - 350)) * attenuation) ** gamma
        G = 0.2 - 0.2*(wavelength - 380) / (430 - 380) 
        B = (1.0 * attenuation) ** gamma
    elif 430 <= wavelength <= 490:
        R = 0.0
        G = ((wavelength - 430) / (490 - 430)) ** gamma
        B = 1.0
    elif 490 <= wavelength <= 510:
        R = 0.0
        G = 1.0
        B = (-(wavelength - 510) / (510 - 490)) ** gamma
    elif 510 <= wavelength <= 580:
        R = ((wavelength - 510) / (580 - 510)) ** gamma
        G = 1.0
        B = 0.0
    elif 580 <= wavelength <= 645:
        R = 1.0
        G = (-(wavelength - 645) / (645 - 580)) ** gamma
        B = 0.0
    elif 645 <= wavelength <= 850:
        attenuation = 0.2 + 0.8 * (850 - wavelength) / (850 - 645)
        R = (1.0 * attenuation) ** gamma
        G = 0.0
        B = 0.0
    elif 850 <= wavelength <= 900:
        attenuation = 0.7 + 0.3 * (850 - wavelength) / (850 - 800)
        R = (1.0 * attenuation) ** gamma
        G = 0.5 - 0.4*(1000 - wavelength) / (1000 - 800)
        B = 0.5 - 0.4*(1000 - wavelength) / (1000 - 800)
    else:
        R = 0.0
        G = 0.0
        B = 0.0
    R *= 255
    G *= 255
    B *= 255

    color = f'#{int(R):02X}{int(G):02X}{int(B):02X}'
    return color

# Get colors for all filters
filters_colors = {}

all_filters = ["SPLUS_U", "SPLUS_F378", "SPLUS_F395", "SPLUS_F410", "SPLUS_F430", "SPLUS_G", "SPLUS_F515", "SPLUS_R", "SPLUS_F660", "SPLUS_I", "SPLUS_F861", "SPLUS_Z",
               "GAIA3_G", "GAIA3_RP", "GAIA3_BP"]
for filt in all_filters:
    lambda_ref = _lambda_ref[filt]
    filters_colors[filt] = wavelength_to_hex(lambda_ref)

filters_colors
```

```python
def mAB_to_fluxwl(mAB, wl):
    """
    Calculates the flux in units of [W nm-1 m-2] from AB magnitudes, for a given wl given in [nm]

    Parameters
    ----------
    mAB : float
        magnitude in the AB system
    wl : float
        wavelength given in [nm]
    
    Returns
    -------
    float
        flux in units of [W nm-1 m-2]

    """

    fwl = 1.088480625 * 10**(-5) * wl**(-2) * 10**(-0.4*mAB)

    return fwl

def fluxwl_to_mAB(flux, wl):
    """
    Calculates the AB magnitudes for the flux, in units of [W nm-1 m-2], for a given wl given in [nm]

    Parameters
    ----------
    flux : float
        in units of [W nm-1 m-2]
    wl : float
        wavelength given in [nm]
    
    Returns
    -------
    float
        magnitude in the AB system

    """

    mAB = -2.5 * np.log10(9.18695452 * 10**(4) * wl**2 * flux)

    return mAB
```

```python
from matplotlib import pyplot as plt
from matplotlib.legend_handler import HandlerTuple
import numpy as np

fig, axs = plt.subplots(2,1, figsize= (12,7), height_ratios=[2, 1])
source = 2642100160642247424

ax = axs[0]

i = np.where(calibrated_spectra["source_id"] == source)[0][0]

norm = np.max(calibrated_spectra.loc[i, "flux"])
ax.plot(sampling*10, calibrated_spectra.loc[i, "flux"]/1e-17, lw = 0.5, color = "#666666", alpha = 1)
ax.fill_between(sampling*10, (calibrated_spectra.loc[i, "flux"]/1e-17-calibrated_spectra.loc[i, "flux_error"]/1e-17), (calibrated_spectra.loc[i, "flux"]/1e-17+calibrated_spectra.loc[i, "flux_error"]/1e-17), color = "#666666", alpha = 0.2)

ax.scatter(sampling*10, calibrated_spectra.loc[i, "flux"]/1e-17, s = 2, lw = 0.5, color = "#666666", alpha = 1)


filters = ["SPLUS_U", "SPLUS_F378", "SPLUS_F395", "SPLUS_F410", "SPLUS_F430", "SPLUS_G", "SPLUS_F515", "SPLUS_R", "SPLUS_F660", "SPLUS_I", "SPLUS_F861", "SPLUS_Z"]
broad = ["SPLUS_U", "SPLUS_G", "SPLUS_R", "SPLUS_I", "SPLUS_Z"]
narrow = ["SPLUS_F378", "SPLUS_F395", "SPLUS_F410", "SPLUS_F430", "SPLUS_F515", "SPLUS_F660", "SPLUS_F861"]

for filt in filters:
    filter_data = pd.read_csv(f"filters/{filt}.dat", delim_whitespace=True)
        
    #Norm = np.nanmax(filter_data.iloc[:,1])    
    # get color for the plot
    color = filters_colors[filt]
    
    # plot the filter
    x = filter_data.iloc[:,0]
    y = 0.5*filter_data.iloc[:,1]*norm/1e-17
    x = np.concatenate(([x[0]], x, [list(x)[-1]]))
    y = np.concatenate(([0], y, [0]))
    
    ax.fill_between(x, 0, y,
        color = color, zorder = 1, alpha = 0.3)
    
    linewidth = 1
    
    ax.plot(x, y,
            color = color, alpha = 0.5, linewidth = linewidth)

# Plot synthetic magnitudes
j = np.where(input_data["source_id"] == source)[0][0]


ax.plot([],[], color = filters_colors["SPLUS_U"], ls = '-', label = "SPLUS_U")

for filt in filters:

    if filt == "SPLUS_U":
        continue
    
    filt_color = filters_colors[filt]
    filt_wl = _lambda_ref[filt]/10

    mAB = input_data[f"{filt}_mod"][j]
    flux = mAB_to_fluxwl(mAB, filt_wl)

    if filt in broad:
        marker = "s"
        size = 100
    elif filt in narrow:
        marker = "^"
        size = 50

    ax.scatter(filt_wl*10, flux/1e-17, color = filt_color, zorder = 10, marker=marker, s = size)

    ax.text(filt_wl*10, flux/1e-17 - 0.2, f"({mAB:.1f})", color = filt_color, ha = "center", va = "top")
    ax.plot([],[], marker = marker, color = filt_color, ls = '-', label = filt)

    #ax.text(filt_wl*10, flux+0.2e-17, f"{filt.replace('SPLUS_', '')}", color = filt_color, ha = "left", va = "bottom")

ax.legend(ncol = 3, fontsize = 8)
ax.fill_between([0,3360],[0,0],[norm/1e-17*1.15,norm/1e-17*1.15], color = "#888888", alpha = 0.4, zorder = -1)
ax.fill_between([10200,15000],[0,0],[norm/1e-17*1.15,norm/1e-17*1.15], color = "#888888", alpha = 0.4, zorder = -1)

ax.text(0.02, 0.93, f"GAIA DR3 source: {source}", transform=ax.transAxes, bbox=dict(facecolor='#FFFFFF', alpha = 0.5, edgecolor='#888888', pad=4.0))

ax.minorticks_on()
ax.set_ylabel(r"Flux [$10^{-17}$ W nm$^{-1}$ m$^{-2}$]")
ax.set_ylim(0, norm/1e-17*1.15)
ax.set_xlim(2500,11000)

ax = axs[1]

ax.minorticks_on()
ax.set_xlabel(r"Wavelength [$\AA$]")
ax.set_ylabel("magnitude")
ax.set_xlim(2500,11000)
ax.set_ylim(18,7)

ax.fill_between([0,3360],[0,0],[30,30], color = "#888888", alpha = 0.4, zorder = -1)
ax.fill_between([10200,15000],[0,0],[30,30], color = "#888888", alpha = 0.4, zorder = -1)

p1, = ax.plot([], [], ls = '', marker = 's', color = "#222222")
p2, = ax.plot([], [], ls = '', marker = '^', color = "#222222")
p3, = ax.plot([], [], ls = '', marker = 's', color = "#FFFFFF", markeredgecolor = "#222222")
p4, = ax.plot([], [], ls = '', marker = '^', color = "#FFFFFF", markeredgecolor = "#222222")

l = ax.legend([(p3, p4), (p1, p2)], ['S-PLUS instrumental magnitudes', 'Gaia XPSP magnitudes [AB-system]'], handler_map={tuple: HandlerTuple(ndivide=None)}, ncols = 2, loc = 9)

k = np.where(inst_df["source_id"] == source)[0][0]

for filt in filters:

    filt_color = filters_colors[filt]
    filt_wl = _lambda_ref[filt]/10

    if filt in broad:
        marker = "s"
        size = 10
    elif filt in narrow:
        marker = "^"
        size = 7

    if filt == "SPLUS_U":
        mAB_inst = inst_df[f"{filt}"][k]
        ax.plot(filt_wl*10, mAB_inst, ls = '', marker = marker, markersize = size, color = "#FFFFFF", markeredgecolor = filt_color)
        

    else:        
        mAB_gaia = input_data[f"{filt}_mod"][j]
        ax.plot(filt_wl*10, mAB_gaia, ls = '', marker = marker, markersize = size, color = filt_color, markeredgecolor = filt_color, zorder = 2)

        mAB_inst = inst_df[f"{filt}"][k]
        ax.plot(filt_wl*10, mAB_inst, ls = '', marker = marker, markersize = size, color = "#FFFFFF", markeredgecolor = filt_color, zorder = 3)        

        ax.plot([filt_wl*10, filt_wl*10], [mAB_gaia, mAB_inst], ls = '--', linewidth = 0.5, color = filt_color, zorder = 1)

        if filt in ["SPLUS_F378"]:
            ax.text(filt_wl*10, mAB_gaia+0.5, f"{mAB_gaia-mAB_inst:.3f}", color = filt_color, ha = "center", va = "top", fontsize = 9)

        elif filt in ["SPLUS_F395", "SPLUS_F410"]:
            ax.text(filt_wl*10, mAB_inst-0.3, f"{mAB_gaia-mAB_inst:.3f}", color = filt_color, ha = "center", va = "bottom", fontsize = 9)

        else:
            ax.text(filt_wl*10 + 20, mAB_inst + (mAB_gaia-mAB_inst)/2 + 0.1, f"{mAB_gaia-mAB_inst:.3f}", color = filt_color, ha = "left", va = "center", fontsize = 9)
# get Gaia spectra in AB magnitude

flux_spectra = calibrated_spectra.loc[i, "flux"]
wl_spectra = sampling
mAB_spectra = fluxwl_to_mAB(flux=flux_spectra, wl=wl_spectra)

ax.plot(wl_spectra*10, mAB_spectra, lw = 0.5, color = "#666666", alpha = 1, zorder = -5)
ax.scatter(wl_spectra*10, mAB_spectra, s = 2, lw = 0.5, color = "#666666", alpha = 1, zorder = -4)

axs[0].tick_params(axis="x",direction="in", pad=-15, which = 'both')
axs[0].axes.xaxis.set_ticklabels([])

plt.tight_layout()
plt.subplots_adjust(hspace = 0)

plt.savefig("gaia_xpsp_example.png", dpi = 90)
```

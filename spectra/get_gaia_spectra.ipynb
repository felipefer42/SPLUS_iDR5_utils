```python
# Current (iDR5)
#http://svo2.cab.inta-csic.es/svo/theory/fps3/index.php?mode=browse
lambda_ref = {  'SPLUS_U': 3533.29, 'SPLUS_F378': 3773.13,
              'SPLUS_F395': 3940.70, 'SPLUS_F410': 4095.27,
              'SPLUS_F430': 4292.39,    'SPLUS_G': 4758.49,
              'SPLUS_F515': 5133.15,    'SPLUS_R': 6251.83,
              'SPLUS_F660': 6613.88,    'SPLUS_I': 7670.59,
              'SPLUS_F861': 8607.59,    'SPLUS_Z': 8936.64,
               'GALEX_FUV': 1535.08,  'GALEX_NUV': 2300.78,
                  'SDSS_U': 3556.52,     'SDSS_G': 4702.50,
                  'SDSS_R': 6175.58,     'SDSS_I': 7489.98,
                  'SDSS_Z': 8946.71,       'PS_G': 4849.11,
                    'PS_R': 6201.20,       'PS_I': 7534.96,
                    'PS_Z': 8674.20,    'GAIA_BP': 5048.62,
                  'GAIA_G': 6246.76,    'GAIA_RP': 7740.87,
                    'SM_U': 3493.36,       'SM_V': 3835.93,
                    'SM_G': 5075.19,       'SM_R': 6138.44,
                    'SM_I': 7767.98,       'SM_Z': 9145.99,
                 'GAIA3_G': 6217.59,   'GAIA3_BP': 5109.71,
                'GAIA3_RP': 7769.02}

# Plot synthetic magnitudes
def wavelength_to_hex(wavelength, gamma=0.8):
    '''This converts a given wavelength of light to an
    approximate hex color value. The wavelength must be given
    in angstrons in the range from 3000 A through 10000 A

    Based on code by Dan Bruton
    http://www.physics.sfasu.edu/astro/color/spectra.html
    '''

    wavelength = float(wavelength) / 10.

    if 200 <= wavelength < 350:
        attenuation = 0.5 + 0.5 * (wavelength - 200) / (440 - 200)
        R = ((-(wavelength - 440) / (440 - 200)) * attenuation) ** gamma
        G = 0.0
        B = (1.0 * attenuation) ** gamma
    elif 350 <= wavelength <= 380:
        attenuation = 0.8 + 0.2 * (wavelength - 350) / (430 - 350)
        R = ((-(wavelength - 430) / (430 - 350)) * attenuation) ** gamma
        G = 0.2*(wavelength - 350) / (380 - 350) 
        B = (1.0 * attenuation) ** gamma
    elif 380 <= wavelength <= 430:
        attenuation = 0.3 + 0.7 * (wavelength - 350) / (430 - 350)
        R = ((-(wavelength - 430) / (430 - 350)) * attenuation) ** gamma
        G = 0.2 - 0.2*(wavelength - 380) / (430 - 380) 
        B = (1.0 * attenuation) ** gamma
    elif 430 <= wavelength <= 490:
        R = 0.0
        G = ((wavelength - 430) / (490 - 430)) ** gamma
        B = 1.0
    elif 490 <= wavelength <= 510:
        R = 0.0
        G = 1.0
        B = (-(wavelength - 510) / (510 - 490)) ** gamma
    elif 510 <= wavelength <= 580:
        R = ((wavelength - 510) / (580 - 510)) ** gamma
        G = 1.0
        B = 0.0
    elif 580 <= wavelength <= 645:
        R = 1.0
        G = (-(wavelength - 645) / (645 - 580)) ** gamma
        B = 0.0
    elif 645 <= wavelength <= 850:
        attenuation = 0.2 + 0.8 * (850 - wavelength) / (850 - 645)
        R = (1.0 * attenuation) ** gamma
        G = 0.0
        B = 0.0
    elif 850 <= wavelength <= 900:
        attenuation = 0.7 + 0.3 * (850 - wavelength) / (850 - 800)
        R = (1.0 * attenuation) ** gamma
        G = 0.5 - 0.4*(1000 - wavelength) / (1000 - 800)
        B = 0.5 - 0.4*(1000 - wavelength) / (1000 - 800)
    else:
        R = 0.0
        G = 0.0
        B = 0.0
    R *= 255
    G *= 255
    B *= 255

    color = f'#{int(R):02X}{int(G):02X}{int(B):02X}'
    return color

# Get colors for all filters
filters_colors = {}

all_filters = ["SPLUS_U", "SPLUS_F378", "SPLUS_F395", "SPLUS_F410", "SPLUS_F430", "SPLUS_G", "SPLUS_F515", "SPLUS_R", "SPLUS_F660", "SPLUS_I", "SPLUS_F861", "SPLUS_Z",
               "GAIA3_G", "GAIA3_RP", "GAIA3_BP"]
for filt in all_filters:
    lambda_ref = _lambda_ref[filt]
    filters_colors[filt] = wavelength_to_hex(lambda_ref)

filters_colors
```

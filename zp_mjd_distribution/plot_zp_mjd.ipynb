```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data_file = "../auxiliary_tables/iDR5_complete_info_fields_table.csv"

iDR5_fields = pd.read_csv(data_file)
iDR5_fields.columns
```

```python
filters = ["U", "F378", "F395", "F410", "F430", "G", "F515", "R", "F660", "I", "F861", "Z"]

f_MAR = iDR5_fields["Reduction"] == "MAR"

filt = "R"
x = iDR5_fields[f"year_{filt}"]
y = iDR5_fields[f"ZP_{filt}"]

fig, ax = plt.subplots(figsize = (10,4))
ax.scatter(x[f_MAR],y[f_MAR], label = "MAR", s = 5)
ax.scatter(x[~f_MAR],y[~f_MAR], label = "JYPE", s = 5)
ax.legend()
```

```python
f_psf = iDR5_fields["Ref_mag_inst"] == "psf"

filt = "F861"
x = iDR5_fields[f"year_{filt}"]
y = iDR5_fields[f"ZP_{filt}"]
y_err = iDR5_fields[f"aper_corr_dual_{filt}"]

fig, ax = plt.subplots(figsize = (10,4))
ax.scatter(x[f_psf],y[f_psf], label = "mag_inst: psf", s = 3)
ax.scatter(x[~f_psf],y[~f_psf], label = "mag_inst: dual", s = 3)
ax.legend()
```
